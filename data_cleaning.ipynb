#### Data cleaning steps for text column - transaction_details

✅ Cleaned & Standardized Categories

1. IndiaForensic Transactions

(All variants grouped)

indiaforensic services
indiaforensic income (monthly)
transfer to indiaforensic
transfer from indiaforensic
indiaforensic bbps / aeps / pos / settlement
indiaforensic merchant / wallet / escrow
indiaforensic charges / fees / recovery

👉 Normalize to:
INDIAFORENSIC_TRANSACTION

2. BBPS (Bharat Bill Payment System)

Variants:

bbps settlement / settlment / setmnt
bbps dt / dtd / dated
bbps ox / xo / reversal
bbps certificate fee / tds

👉 Normalize to:
BBPS_SETTLEMENT

3. AEPS (Aadhaar Enabled Payment System)

Variants:

aeps income / gst / fees
aeps acquiring / bank income
aeps npci withdrawal / deposit
aeps reversal entries

👉 Normalize to:
AEPS_TRANSACTION

4. NEFT Transactions

Huge volume (names + banks mixed)

👉 Normalize:

neft + bank + name

Example:

neftsbinrahul → NEFT_TRANSFER_SBI
neftaxisjohn → NEFT_TRANSFER_AXIS

👉 Final tag:
NEFT_TRANSFER

(Optional sub-tag by bank)

5. RTGS Transactions

Variants:

rtgs + bank
rtgs charges

👉 Normalize:
RTGS_TRANSFER

6. IMPS Transactions

Variants:

imps ow / iw
imps return / penalty / chargeback

👉 Normalize:
IMPS_TRANSACTION

Subtypes:

IMPS_OUTWARD
IMPS_INWARD
IMPS_RETURN
IMPS_CHARGEBACK
7. POS / Card Settlements

Variants:

visa pos acq
master / maestro / rupay
pos settlement / reversal

👉 Normalize:
CARD_SETTLEMENT

8. Cash Deposits / Pickup / DSB

Variants:

cashdep*
dsb cash pickup
beat cash pickup
excess / short cash

👉 Normalize:
CASH_TRANSACTION

Subtypes:

CASH_DEPOSIT
CASH_PICKUP
CASH_DISCREPANCY
9. Charges / Fees / Taxes

Variants:

service tax / gst / cess
commission charges
processing fee
penalty

👉 Normalize:
BANK_CHARGES

10. Internal Transfers

Variants:

internal fund transfer
ib tpft / oft
transfer to self

👉 Normalize:
INTERNAL_TRANSFER

11. Vendor / Utility Payments

Examples:

airtel / vodafone / mtnl
insurance companies
electricity boards
tata / airtel etc.

👉 Normalize:
VENDOR_PAYMENT

12. Foreign / Forex Transactions

Variants:

usd payments
inward/outward remittance
firc / fx charges

👉 Normalize:
FOREX_TRANSACTION

13. Cheque Transactions

Variants:

chq deposit / return
inward clearing

👉 Normalize:
CHEQUE_TRANSACTION

14. Loan / Interest / FD

Variants:

fd booking
loan recovery
interest

👉 Normalize:
LOAN_FD_TRANSACTION
